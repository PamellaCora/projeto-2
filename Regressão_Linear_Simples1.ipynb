## Importando conjunto de dados

```python
import pandas as pd
```

```python
dados = pd.read_csv("C:/Users/20221enpro0154/Downloads/Salary_Data.csv")
#Não precisa colocar separador nem decimal, pois o separador de decimal é ponto e os dados separados por vírgula (padrão internacional)
dados.head #ler a cabeça, os primeiros dados
dados #aparece todos
```

## Gráficos

```python
import matplotlib.pyplot as plt #importou só a função (colocou biblioteca . alguma coisa)
```

```python
# Diagrama de dispersão
plt.scatter(dados['YearsExperience'], dados['Salary'], color = "darkkhaki")
            #O primeiro é o que fica no X, o que normalmente é a causa, ou seja, anos de experiencia estão levando aos salários maiores
plt.title("Salário vs Experiência")
plt.xlabel("Anos de Experiência")
plt.ylabel("Salário (Doláres americanos por ano)")
```

```python
#Modelo de Regressão
#Primeiro separar o X e Y
x = dados.iloc[:,:1] #Usar todas as linhas é usar dois pontos e separar por vírgula (Antes da vírgula são as linhas e depois da vírgula são as colunas. Depois do segundo dois pontos ele pega tudo menos o valor que está nele)
y = dados.iloc[:,1:] #Pegar a coluna 1 até o final, que é só a coluna 1 neste exemplo
```

```python
x #mostrando x
```

```python
y #mostrando Y
```

```python
# Modelo

from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

modelo = LinearRegression()
modelo.fit(x,y) #fit é regular, então ele ta fazendo o B0 e B1
```

```python
#Coeficientes do modelo (Betas)
print('Coeficiente angular (beta_1): ', modelo.coef_)
print('Intercepto (beta_2): ', modelo.intercept_)
```

```python
# Gerando predições para os próprios valores de Y

y_pred  = modelo.predict(x)
```

```python
#Gerando gráfico com a reta
plt.scatter(x, y, color = "hotpink")
plt.plot (x, y_pred, color = "darkkhaki") #diagrama de dispersão com a reta
plt.title ("Salário vs Experiência")
plt.xlabel ("Anos de Experiência")
plt.ylabel ("Salário anual em dólares americanos")
```

```python
# Coeficiente de determinação (R²)
r2_score (y, y_pred) # Para saber se há correlação entre as variáveis)
```
